```python
df = df[df['Type Of Policy'] != 'LT']

df = df[df['Type Of Policy'] != 'PENSION INDIVIDUAL']

df = df[df['Type Of Policy'] != 'ANNUITY']

df.head()
```

### Removing Upendo Products

```python
df.columns
```

```python
# Defining the Specific Column using Regex:

upendo_format = r'\d{6}/\d{2}'

# Remove Rows where 'Policy Number' is an Upendo Product

df = df[~df['Policy Number'].str.match(upendo_format)]

df.head()
```

### Removing Duplicate Rows

```python
df.columns
```

```python
# Removing Duplicated Policy Numbers

df = df.drop_duplicates(subset='Policy Number', keep='first')

df.head()
```

```python
# Confirm there are no Duplicates in the Policy Numbers

are_duplicates = df['Policy Number'].duplicated().any()

if are_duplicates:
    print('Duplicates Present')
else:
    print('Duplicates Absent')
```

### Aligning WealthBuidler Premiums

```python
df.columns
```

```python
# Identify rows that match the specified regex pattern

regex_pattern = r'^WBD/([A-Za-z0-9]+)/\d{5,8}/\d{2,4}$'

matching_rows = df['Policy Number'].str.contains(regex_pattern)

# Count the number of rows that match the specified regex pattern in the 'Policy Number' column

matching_rows_count = (df['Policy Number'].str.contains(regex_pattern)).sum()


# Display the count of rows that match the regex pattern

print("\nNumber of rows matching the regex pattern:", matching_rows_count)
```

```python
# Define the regex pattern

regex_pattern = r'^WBD/([A-Za-z0-9]+)/\d{5,8}/\d{2,4}$'

# Identify rows that match the specified regex pattern in the 'Policy Number' column

mask = df['Policy Number'].str.contains(regex_pattern)

# Convert string-formatted numeric values to actual numeric values

df['Booked Premium'] = pd.to_numeric(df['Booked Premium'].astype(str).str.replace(',', ''), errors='coerce')

df['Total '] = pd.to_numeric(df['Total '].astype(str).str.replace(',', ''), errors='coerce')

df['Annualized '] = pd.to_numeric(df['Annualized '].astype(str).str.replace(',', ''), errors='coerce')

# Duplicate 'Booked Premium' values to 'Annualized' and 'Total' columns for the matching rows

df.loc[mask, ['Annualized ', 'Total ']] = df.loc[mask, 'Booked Premium'].repeat(2).values.reshape(-1, 2)

df.head()
```
